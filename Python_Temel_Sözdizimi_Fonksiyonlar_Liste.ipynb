```python
#def=definition
#dairealanıhesapla=fonksiyon ismi
#r=girdi parametresi
def dairealanhesapla(r):
#dairenin alanını hesaplar girdi:yarıçapp çıktı=dairenin alanı

  pi=3.14
  daire_alanı=pi*(r**2)
  return daire_alanı
dairealanhesapla(5)
```

```python
def cemberincevresihesapla(r):
#çemberin çevresi hesaplar girdi=yarıçap çıktı=çemberin çevresi
  pi=3.14
  cember_cevresi_cevresi=2*pi*r
  return cember_cevresi
cemberincevresihesapla(6)

```
