**lambda fonksiyonu**:ileri seviye bir fonk. her hangi bir sayıda argüman alabilir ancak yalnızca bir ifadeye sahip olabilir.



```python
#çarpma işlemi yapan normal fonk.
def carpma (x,y,z):
  return x*y*z
print(carpma(2,3,4))
```

```python
fonksiyon_lambda=lambda x,y,z:x*y*z
fonksiyon_lambda(2,3,4)
```

lambda fonk onları başka bir fonk içinde anonim bir fonk olarak kullandığımızda daha iyi kullanıllır.

```python
def fonksiyonLambda(n):
  return lambda a:a*n
ana_fonksiyon=fonksiyonLambda(3)
print(ana_fonksiyon(5)) # Lambda fonksiyonunu çağırmanın doğru yolu
```

bu kod bir "çarpan fabrikası" gibi çalışır. fonksiyonLambda ile istediğimiz katsayıda çarpan üreten fonksiyonlar yaratabiliriz (örneğin 3'le çarpan, 5'le çarpan vb.).
