KOŞULLU İFADELER: bir koşul tanımlayıp o koşulun doğru veya yanlış  değerlendirilmesine bağlı olarak hesaplamalar gerçekleştiren özellik.

```python
#büyük küçük karşılaştırılması
sayi1=12.0
sayi2=20.0

if sayi1 < sayi2:
  print("sayi1 küçüktür sayi2")
elif sayi1 > sayi2:
  print("sayi1 büyüktür sayi2")
else: print("sayi1 eşittir sayi2")
```
